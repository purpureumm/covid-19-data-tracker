## Dataset Description

For this project, three datasets published by the United States Centers for Disease Control and Prevention (CDC) are used. All three dataset is being continuously updated by the CDC daily.

The first dataset used in the project is “COVID-19 Vaccinations in the United States, Jurisdiction”. This dataset contains the overall US COVID-19 vaccines deliveries and administration data at national and jurisdiction level. For the purpose of this project, the jurisdictions that are of interest are the 50 U.S. States. There is a total of 80 variables included in the dataset. We will be only using the following variables for this project: Date, Location, Administered_Dose1_Pop_Pct, and Series_Complete_Pop_Pct. These four variables describe the date of the data, location of the data, percent of the population that has received one dose of any COVID-19 vaccines, and percent of the population that have received two doses (fully vaccinated) of any COVID-19 vaccines.

The second dataset used is “COVID-19 Vaccinations in the United States, County”. This dataset contains the overall US COVID-19 vaccine administration and vaccine equity data at county level. There is a total of 32 variables in this dataset. We will be only using five variables from this dataset for the project. Variables Date, Recip_County, Recip_State, Administered_Dose1_Pop_Pct, and Series_Complete_Pop_Pct will be used. These variables contain the date, county name, state name, and percent of the population that has received one dose of any COVID-19 vaccines, and percent of the population that have received two doses (fully vaccinated) of any COVID-19 vaccines.

The third dataset used is “United States COVID-19 County Level of Community Transmission as Originally Posted”. This dataset contains the community transmission levels for all available counties. There are 7 variables in this dataset, but we will be using only five out of the seven variables: state_name, county_name, report_date, cases_per_100K_7_day_count_change, and percent_test_results_reported_positive_last_7_days. The last two variables are the total number of new cases per 100,000 persons within the last 7 days and the percentage of positive COVID-19 tests during the last 7 days.

##  Question of Interest

The purpose of this project is to develop a web app using Dash, a framework built on Python by plotly for making interactive web applications. The goal of this app is to help visualize the COVID-19 data gathered by the CDC to anyone interested in learning more about the spread of COVID-19 in the U.S. and the vaccination effort against it. The app aims to use the interactive graphics and tables built into Dash to give the user a clear picture of the COVID-19 data we chose, as well as using the interactive feature to allow the user to select parts of the data that are they are interested in.

The web app will include 3 interactive graphs and 1 table. Graph 1 will display the map of the U.S where the states are colored by the vaccination status (1 dose or fully vaccinated)  up to the date selected. The table will display the vaccination status of the counties in a selected state up to the date chosen. The last two graphs will display the trend of any period of choice of the 7-day moving average of new cases per 100,000 people and the percentage of positive COVID-19 tests, filtered by county and state.
